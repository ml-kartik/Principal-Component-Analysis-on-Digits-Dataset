```python
import numpy as np
np.unique(digits.target)
```

```python
df = pd.DataFrame(digits.data, columns = digits.feature_names)
df.head()
```

```python
df.describe() # 0 is min and 16 is max value
```

```python
X = df
y = digits.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2, random_state = 30)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
Xtrain_scaled = scaler.fit_transform(Xtrain)
Xtest_scaled = scaler.transform(Xtest)
```

### Without PCA Score

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(Xtrain_scaled,ytrain)
model.score(Xtest_scaled,ytest)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95) # n_components = 0.95 means retain 95% of useful information
```

#### Before PCA shape

```python
Xtrain_scaled.shape
```

```python
Xtrain_pca = pca.fit_transform(Xtrain_scaled)
Xtest_pca = pca.transform(Xtest_scaled)
```

#### After PCA Shape

```python
Xtrain_pca.shape
```

### With PCA Score

```python
pca_model = LogisticRegression()
pca_model.fit(Xtrain_pca,ytrain)
pca_model.score(Xtest_pca,ytest)
```

## Accuracy drops because pca drops dimenstion/Indepedent Fetrures which are not important for predictions
